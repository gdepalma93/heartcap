```python
# Import Dependencies
import pandas as pd
import numpy as np
import datetime
import pandas_datareader.data as web
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
start = datetime.datetime(2010,1,1)
end = datetime.datetime(2020,4,30)

# Declare Yahoo Symbols
btc = "BTC-USD"
eth = "ETH-USD"
xlm = "XLM-USD"

# Scrape Yahoo data
df_btc = web.DataReader(btc, 'yahoo', start, end)
df_eth = web.DataReader(eth, 'yahoo', start, end)
df_xlm = web.DataReader(eth, 'yahoo', start, end)

dfs = [df_btc, df_eth, df_xlm]

# calculate rolling stats
# Rolling Mean   
close_px_btc = df_btc['Adj Close']
close_px_eth = df_eth['Adj Close']
close_px_xlm = df_xlm['Adj Close']

# Bitcoin moving_avg
mavg_btc = close_px_btc.rolling(window=100).mean()
mavg_eth = close_px_eth.rolling(window=100).mean()
mavg_xlm = close_px_xlm.rolling(window=100).mean()
```

```python
## Bitcoin Timeseries

#Visualization
%matplotlib inline

# Adjusting the size of matplotlib
import matplotlib as mpl
mpl.rc('figure', figsize=(8,7))
mpl.__version__

# Adjusting the style of matplotlib
style.use('ggplot')

close_px_btc.plot(label='BTC')
mavg_btc.plot(label='mavg')
plt.legend()

# Plot Returns
# plot Returns
rets = close_px_btc / close_px_btc.shift(1) - 1

rets.plot(label='return')
```

```python
# Plot Returns
rets = close_px_btc / close_px_btc.shift(1) - 1

rets.plot(label='return')
```

```python
# Plot cumulative returns if invested from start date
(rets+1).cumprod().plot()
```
